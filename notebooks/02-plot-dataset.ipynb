```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_theme()
```

```python
df = pd.read_csv('data/mec-dataset.csv')
```

```python
df = df['cohesion thematic_coherence formal_register text_typology'.split()]
df
```

```python
freq = df.apply(pd.Series.value_counts)
freq
```

```python
freq.columns = 'Coesão', 'Coerência Temática', 'Registro Formal', 'Tipologia Textual'
freq.index = [f'Nível {t}' for t in freq.index]
freq
```

```python
plt.figure(dpi=300, figsize=(12,6))
ax = freq.plot.bar(rot=0, ax=plt.gca(), ylabel='Exemplos')
ax.figure.savefig('data/value_counts.pdf')
```
