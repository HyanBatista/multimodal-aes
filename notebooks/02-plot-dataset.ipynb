```python
import pandas as pd
import seaborn as sns
import os
sns.set_theme()
OUTPUT_DIR = 'data/value_counts'
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
df = pd.read_csv('data/mec-dataset.csv')
```

```python
def plot_and_save_value_counts(df, column, label):
    df[column].value_counts().sort_index().plot.bar(xlabel=label, ylabel='Exemplos', rot=0).figure.savefig(f'{OUTPUT_DIR}/{column}.pdf')
```

```python
plot_and_save_value_counts(df, 'cohesion', 'Coes√£o')
```

```python
plot_and_save_value_counts(df, 'formal_register', 'Registro Formal')
```

```python
plot_and_save_value_counts(df, 'text_typology', 'Tipologia Textual')
```

```python
plot_and_save_value_counts(df, 'thematic_coherence', 'Coer√™ncia Tem√°tica')
```
